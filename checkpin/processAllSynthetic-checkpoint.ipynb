```python
import scipy.signal, scipy.io, os
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import FastICA
%matplotlib inline
```

```python
def plotDataSection(data, title, 
    startTime=0, 
    endTime=None, 
    sampleRate=24e3,
    idxPeakMarker=np.array([]),
    fig=None,
    ax=None,
    legend=False
):
    '''
    plot data in a certain interval.
    input:
    - data: [idx of channels, idx of datapoints]
    '''
    if data.ndim == 1:
        data = data.reshape(1, -1)
    if endTime is None:
        endTime = data.shape[-1]/sampleRate
        sectionLength = data.shape[-1]
        dataSection = data
    else:
        idxStart = int(startTime*sampleRate)
        idxEnd = int(endTime*sampleRate)
        sectionLength = idxEnd - idxStart
        dataSection = data[:, idxStart: idxEnd]
    time = np.linspace(startTime, endTime, num=sectionLength)
    if ax == None and fig == None:
        fig, ax = plt.subplots()
    for idxCh, chData in enumerate(dataSection):
        ax.plot(time, chData, label=f'Ch {idxCh}')
        if idxPeakMarker.any():
            ax.plot(time[idxPeakMarker], chData[idxPeakMarker], '*')
    ax.set_title(title)
    ax.set_xlim([startTime, endTime])
    ax.set_xlabel('time [s]')
    ax.set_ylabel('voltage [uv]')
    ax.grid(True)
    if legend:
        ax.legend()
    return fig, ax

def plotSectionMeanAndSTD(data, title, sampleRate=24e3):
    dataMean = data.mean(axis=0)
    time = np.linspace(0, len(dataMean)/sampleRate, num=len(dataMean))
    dataSTD = data.std(axis=0)
    fig = plt.figure(figsize=(10, 10))
    ax1 = fig.add_subplot(211)
    ax1.plot(time, dataMean, color='b')
    ax1.plot(time, dataMean + dataSTD, 'r.')
    ax1.plot(time, dataMean - dataSTD, 'r.')
    ax1.set_title(title)
    ax1.set_xlabel('time [s]')
    ax1.set_ylabel('voltage [uv]')
    ax1.grid(True)
    ax2 = fig.add_subplot(212)
    ax2.plot(time, dataSTD)
    ax2.set_xlabel('time [s]')
    ax2.set_ylabel('STD')
    ax2.grid(True)

def getPeakSections(data, idxPeaks, windowSizeBefore, windowSizeAfter):
    peakSections = []
    for i in idxPeaks:
        peakSections.append(data[i - windowSizeBefore: i + windowSizeAfter])
    return np.array(peakSections)

def removeOutOfBoundIdx(idx, lowBound, upBound):
    '''Remove the idx that are too close to the begining or end 
    defined by lowBound and upBound.
    Needed to avoid index out of bound error when cutting data into small sections.
    '''
    return idx[np.logical_and(idx > lowBound, idx < upBound)]

def plotPSD(data, sampleRate, title):
    freq, PSD = scipy.signal.welch(data, sampleRate, window=scipy.signal.windows.hann(512))
    fig, ax = plt.subplots()
    ax.plot(freq, PSD.transpose())
    ax.set_xlim([0, 150])
    ax.set_yscale('log')
    ax.set_xlabel('frequency [Hz]')
    ax.set_ylabel('PSD')
    ax.set_title(title)
    ax.grid(True)
    return fig, ax
```

Load data

```python
sampleRate = 300
dataFolder = 'Synthetic_Data'
fileList = os.listdir(dataFolder)
dataset = []
for file in fileList:
    data = scipy.io.loadmat(dataFolder + '/' + file)
    dataset.append(data['val'].squeeze())
```

Parameters

```python
filterOrder = 2
bandpassWidth = 1.5
welchWindowSize = 512
highpassFreq = 3.17
lowpassFreq = 50
tukeyAlpha = 0.1
maxIter = 300
maxBPM = 250
minBPM = 30
intervalSTD_threshold = 0.18
```

Processing functions

```python
def filterData(data, 
    sampleRate, 
    filterOrder,
    bandpassWidth, 
    welchWindowSize,
    highpassFreq,
    lowpassFreq,
    tukeyAlpha,
):
    # ==== band stop filter ====
    freq, PSD = scipy.signal.welch(data, sampleRate, 
        window=scipy.signal.windows.hann(welchWindowSize))
    spikeFreq = freq[np.argmax(np.sum(PSD, axis=0))]
    bandstopSOS = scipy.signal.butter(
        filterOrder, (spikeFreq - bandpassWidth/2, spikeFreq + bandpassWidth/2),
        'bandstop', fs=sampleRate, output='sos')
    bandstopfiltData = scipy.signal.sosfiltfilt(bandstopSOS, data, axis=1)
    # plotPSD(bandstopfiltData, sampleRate, 'PSD after band stop filtering')
    # plotDataSection(bandstopfiltData, 'data after high pass', sampleRate=sampleRate).

    # ==== high pass filter ====
    highpassSOS = scipy.signal.butter(filterOrder, highpassFreq, 'highpass', 
        fs=sampleRate, output='sos')
    highpassfiltData = scipy.signal.sosfiltfilt(highpassSOS, bandstopfiltData, axis=1)

    # ==== low pass filter ====
    lowpassSOS = scipy.signal.butter(filterOrder, lowpassFreq, 'lowpass', 
        fs=sampleRate, output='sos')
    lowpassfiltData = scipy.signal.sosfiltfilt(lowpassSOS, highpassfiltData, axis=1)

    # ==== Tukey window ====
    recLength = data.shape[-1]
    tWin = scipy.signal.windows.tukey(recLength, alpha=tukeyAlpha)
    processedData = lowpassfiltData*tWin
    return processedData

def applyICA(processedData, maxIter):
    numCh = processedData.shape[0]
    transform = FastICA(n_components=numCh, whiten='unit-variance', max_iter=maxIter)
    ICAData = transform.fit_transform(processedData.T).T
    return ICAData

def getValidCh(ICAData, sampleRate, minBPM, maxBPM):
    distanceThreshold = sampleRate*60/maxBPM
    heightThreshold = np.percentile(np.abs(ICAData).flatten(), 99)*0.8
    # print(f'height threshold: {heightThreshold}')

    dataset = {}
    dataset['BPM'] = []
    dataset['intervalSTD'] = []
    dataset['data'] = []
    dataset['idxPeaks'] = []
    for chData in ICAData:
        idxPeaks, peakProperties = scipy.signal.find_peaks(
            np.abs(chData), height=heightThreshold, distance=distanceThreshold)
        # minimum number of spikes for the duration of the signal
        minSpikeCount = chData.shape[-1]/sampleRate/60*minBPM 
        if idxPeaks.size > minSpikeCount:
            interval = (idxPeaks[1:] - idxPeaks[:-1])/sampleRate # [seconds]
            intervalSTD = np.std(interval)
            avgInterval = interval.mean() # [seconds]
            avgBPM = 1/avgInterval*60
            if avgBPM > minBPM and avgBPM < maxBPM:
                if chData[idxPeaks].mean() < 0:
                    chData = -chData
                dataset['BPM'].append(avgBPM)
                dataset['intervalSTD'].append(intervalSTD)
                dataset['data'].append(chData)
                dataset['idxPeaks'].append(idxPeaks)

    numValid = len(dataset['BPM'])
    if numValid != 2:
        Error = True
        # print(f'Error! {numValid} valid channels found.')
    else:
        Error = False
    result = dataset
    return Error, result

def plotDataset(dataset, idxTrial, suggestion, sampleRate, childIdx=None):
    numCh = len(dataset['BPM'])
    # childIdx = np.argmax(dataset['BPM'])
    fig, ax = plt.subplots(numCh, 1, figsize=(8, 8))
    fig.tight_layout(h_pad=5)
    for idxCh in range(numCh):
        BPM = dataset['BPM'][idxCh]
        chData = dataset['data'][idxCh]
        idxPeaks = dataset['idxPeaks'][idxCh]
        if idxCh == childIdx or childIdx == None:
            title = f'trial: {idxTrial}, BPM: {BPM:.2f}, {suggestion}'
        else:
            title = f'trial: {idxTrial}, BPM: {BPM:.2f}'
        plotDataSection(chData, title, sampleRate=sampleRate, 
            idxPeakMarker=idxPeaks, fig=fig, ax=ax[idxCh])
```

Main loop

```python
allValidTrial = []
allMotherIntSTD = []
allChildIntSTD = []
for idxTrial, data in enumerate(dataset):
    processedData = filterData(
        data,
        sampleRate,
        filterOrder,
        bandpassWidth, 
        welchWindowSize,
        highpassFreq,
        lowpassFreq,
        tukeyAlpha,
    )
    ICAData = applyICA(processedData, maxIter)
    Error, validDataset = getValidCh(
        ICAData,
        sampleRate,
        minBPM,
        maxBPM,
    )
    # visualize errors
    if Error:
        print(f'Trial {idxTrial}: {len(validDataset["BPM"])} valid channels found.')
        plotDataset(validDataset, idxTrial, 'error', sampleRate)
        continue
    
    montherIdx = np.argmin(validDataset['BPM'])
    childIdx = np.argmax(validDataset['BPM'])
    allValidTrial.append(idxTrial)
    allMotherIntSTD.append(validDataset['intervalSTD'][montherIdx])
    allChildIntSTD.append(validDataset['intervalSTD'][childIdx])

    if validDataset['intervalSTD'][childIdx] > intervalSTD_threshold:
        plotDataset(validDataset, idxTrial, 'child irregular heart rate', sampleRate, childIdx)

    # print(f'Trial {idxTrial}: 2 valid channels found.')
    # print(f'    Ch0, BPM: {validDataset["BPM"][0]}, interval STD: {validDataset["intervalSTD"][0]}')
    # print(f'    Ch1, BPM: {validDataset["BPM"][1]}, interval STD: {validDataset["intervalSTD"][1]}')
```
